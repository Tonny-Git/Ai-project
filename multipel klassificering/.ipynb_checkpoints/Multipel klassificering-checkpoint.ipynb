# SVM / Naive Bayes

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("Data/winequality-red.csv")
```

```python
dataset.head(10)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
X.shape
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_test.shape, y_test.shape)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

## Training the SVM model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', gamma=1, decision_function_shape='ovo', random_state=0)
classifier.fit(X_train, y_train)
```

```python
# Kernal scores:
# linear = 0.6, rbf = 0.69375, poly 0.65
# rbf verkar vara bäst
classifier.score(X_test, y_test)
```

## Training the Naive Bayes model on the Training set

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
```

## Predicting a new result

## Predicting the SVM Test set results

```python
y_pred = classifier.predict(X_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
df = pd.DataFrame({'Orginal value': y_test.flatten(), 'Predicted value': y_pred.flatten()})
df
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## Predicting the Naive Bayes Test set results

```python
y_nb_pred = nb_classifier.predict(X_test)
```

```python
print(np.concatenate((y_nb_pred.reshape(len(y_nb_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
df = pd.DataFrame({'Orginal value': y_test.flatten(), 'Predicted value': y_nb_pred.flatten()})
df
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_nb_pred)
```
